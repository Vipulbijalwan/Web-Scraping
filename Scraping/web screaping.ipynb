### Extracting text from the tags in web page using Python

```python
lines= soup.find_all("p")
```

```python
lines
```

```python
for l in lines:
    print(l.text)
```

Extrecting links in web page

```python
for i in soup.find_all("a"):
    print(i.get("href"))
    
```

Extrecting image information
